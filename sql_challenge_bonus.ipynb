```python
#Importing dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

#connect to the postgresql database
engine = create_engine('postgresql://postgres:postgres@localhost:5432/sql_challenge')
connection = engine.connect()

result = connection.execute('SELECT  title, CAST(ROUND(AVG(salary),2) AS INTEGER) AS Numeric from employees inner join titles on employees.emp_title_id = titles.title_id inner join salaries on employees.emp_no = salaries.emp_no group by title;')
title = []
avgsalary = []
for row in result:
    title.append(row[0])
    avgsalary.append(row[1])
x_axis = np.arange(len(title))
tick_locations = [value for value in x_axis]
plt.bar(title,avgsalary,color ='b',align='center')
plt.xticks(tick_locations, title, rotation="vertical")
plt.show()
#avgsalary
```
